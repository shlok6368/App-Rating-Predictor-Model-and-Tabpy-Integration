# Data Collection - PlayStore Data

```python
import play_scraper
import pandas as pd
import time
collection =  ["TOP_FREE","TOP_PAID","TRENDING"]
category = ['ANDROID_WEAR','ART_AND_DESIGN','AUTO_AND_VEHICLES','BEAUTY','BOOKS_AND_REFERENCE','BUSINESS','COMICS',
            'COMMUNICATION','DATING','EDUCATION','ENTERTAINMENT',
            'EVENTS','FAMILY','FAMILY_ACTION','FAMILY_BRAINGAMES','FAMILY_CREATE','FAMILY_EDUCATION',
            'FAMILY_MUSICVIDEO','FAMILY_PRETEND','FINANCE','FOOD_AND_DRINK','GAME','GAME_ACTION','GAME_ADVENTURE',
            'GAME_ARCADE','GAME_BOARD','GAME_CARD','GAME_CASINO','GAME_CASUAL','GAME_EDUCATIONAL','GAME_MUSIC',
            'GAME_PUZZLE','GAME_RACING','GAME_ROLE_PLAYING','GAME_SIMULATION','GAME_SPORTS','GAME_STRATEGY',
            'GAME_TRIVIA','GAME_WORD','HEALTH_AND_FITNESS','HOUSE_AND_HOME','LIBRARIES_AND_DEMO','LIFESTYLE',
            'MAPS_AND_NAVIGATION','MEDICAL','MUSIC_AND_AUDIO','NEWS_AND_MAGAZINES','PARENTING',
            'PERSONALIZATION','PHOTOGRAPHY','PRODUCTIVITY','SHOPPING','SOCIAL','SPORTS','TOOLS','TRAVEL_AND_LOCAL',
            'VIDEO_PLAYERS','WEATHER']
```

```python
df_dev_info = pd.DataFrame()
for category in category:
    for collection in ["TOP_FREE","TOP_PAID","TRENDING"]:
        time.sleep(2)
        for i in range(1,6):
            try:
                df_dev_info = df_dev_info.append(pd.DataFrame(play_scraper.collection(
                collection=collection,
                category = category,
                results=min(120,int(500/i)),
                page=i-1,)))
            except:
                print("There was some error in fetching collection = ",collection)
                continue
        print(i, "pages ", "for category = ",category, "under collection = ", collection," completed")
            
```

```python
df_dev_info.drop_duplicates(inplace=True)
df_dev_info.to_csv("df_dev_info.csv")
df_dev_info.shape
```

```python
df_dev_info = pd.read_csv("df_dev_info.csv", encoding = 'latin1')
df_dev_info.shape
```

```python
df_app_info_list = []
i = 0
for app_id in list(df_dev_info['app_id']):
    try:
        df_app_info_list.append(play_scraper.details(app_id))
        i=i+1
        if (i%100 == 0):
            print(i,"apps collected out of",df_dev_info.shape[0])
    except:
        continue
```

```python
df_app_info = pd.DataFrame(df_app_info_list)
df_app_info.to_csv("df_app_info.csv")
```
